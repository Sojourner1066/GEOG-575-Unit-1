```python
import json
import csv
```

```python
data_file = open('WB_electricity_access.geojson')
data = json.load(data_file)
data_file.close()
```

```python
data['features'][0]['properties']
```

```python

country_data = []
with open('country.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        country_data.append(row)
```

```python
country_data[0]
```

```python
for i in range(0,len(data['features'])-1):
    for row in country_data:
        if data['features'][i]['properties']['SOV_A3'] == row[2]:
            print(data['features'][i]['properties']['SOV_A3'],row[6])
            data['features'][i]['properties']['region'] = row[5]
            data['features'][i]['properties']['sub-region'] = row[6]
            
```

```python
with open('data.json', 'w') as outfile:
        json.dump(data, outfile)
```
